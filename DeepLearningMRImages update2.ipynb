```python
import numpy as np
import matplotlib.pyplot as plt
import glob
import os
import cv2
import math

# 함수 정의
def remove_extreme_percentiles(data, lower=10, upper=90):
    lower_percentiles = np.percentile(data, lower, axis=1)
    upper_percentiles = np.percentile(data, upper, axis=1)
    mask = [(data[i] < lower_percentiles[i]) | (data[i] > upper_percentiles[i]) for i in range(data.shape[0])]
    return np.array([row[~mask_row] for row, mask_row in zip(data, mask)], dtype=object)

def nearest_square(n):
    return math.ceil(n**0.5)**2

# 대상 디렉토리 설정
base_dir = r'C:\Users\kwoos\Desktop\cording\good\ADND\axial'
method1 = r'C:\Users\kwoos\Desktop\cording\good\Method1\axial'

# 대상 디렉토리 내의 모든 .npy 파일을 찾습니다.
files = glob.glob(os.path.join(base_dir, '*.npy'))

# 각 .npy 파일에 대해 처리 작업을 수행합니다.
for file in files:
    print(f"Processing file: {file}")  # 현재 처리 중인 파일 이름 출력

    # 파일명에서 정보를 가져오기
    file_name = os.path.basename(file)  # 파일 이름과 확장자를 가져옵니다.
    file_base_name = os.path.splitext(file_name)[0]  # 파일 확장자를 제거
    file_info = file_base_name.split('_')  # "_"로 나누어 정보 부분 가져오기
    file_number = file_info[1]  # 번호 부분
    file_type = file_info[-2]  # 종류 부분(sagittal, coronal, axial)

    # 'OAS1_{번호}_{종류}' 폴더를 생성
    image_dir = os.path.join(method1, f'OAS1_{file_number}_{file_type}')
    os.makedirs(image_dir, exist_ok=True)

    # .npy 파일 불러오기
    data = np.load(file)

    # 결과를 저장할 리스트 초기화
    result = []

    # 반복 범위를 데이터의 인덱스 범위 내로 설정
    start = max(0, data.shape[0] // 2 - 80)
    end = min(data.shape[0], data.shape[0] // 2 + 80)

    # i 값을 start부터 end까지 바꿔가며 실행
    for i in range(start, end+1):
        # 'i'가 가운데 인덱스인 경우를 처리
        if i == data.shape[0] // 2:
            continue  # i가 가운데 인덱스일 때는 건너뛰기

        # 데이터가 4D 라면, 첫 번째 축을 따라 i번째 slice를 선택합니다.
        if len(data.shape) == 4:
            data_slice = data[i, :, :, 0]  # 4차원을 2차원으로 줄입니다.

            # 결과 리스트에 추가
            result.append(data_slice)

    # 결과를 numpy array로 변환
    result = np.array(result)

    reshaped_data = result.reshape(160, -1)
    filtered_array = remove_extreme_percentiles(reshaped_data, lower=10, upper=90)

    entropy_elements = {}
    for idx, row in enumerate(filtered_array):
        if row.size != 0:
            image = row
            side_length = int(math.sqrt(nearest_square(row.size)))
            padding = side_length**2 - row.size
            padded_row = np.pad(row, (0, padding), mode='constant')
            image = padded_row.reshape((side_length, side_length))
            image = np.uint8(image)
            hist = cv2.calcHist([image], [0], None, [256], [0,256])
            hist /= hist.sum()  # normalize
            entropy = -np.sum(hist*np.log2(hist + np.finfo(float).eps))
            entropy_elements[idx] = entropy

    filtered_entropy_elements = {k: v for k, v in entropy_elements.items() if int(v) < 50}
    sorted_indices = [key for key, value in sorted(filtered_entropy_elements.items(), key=lambda item: item[1], reverse=True)]
    top_20_indices = sorted_indices[:20]

    # 파일 이름에서 환자 ID를 추출합니다.
    base_name = os.path.basename(file)  # 파일 이름과 확장자를 가져옵니다.
    patient_id = os.path.splitext(base_name)[0].split('_')[1]  # 확장자를 제거하고 '_'로 분할한 뒤 환자 ID를 가져옵니다.

    # 배열에 저장된 인덱스에 해당하는 이미지를 출력하고 저장합니다.
    for idx, i in enumerate(top_20_indices):
        # 현재 이미지를 선택합니다.
        image = result[i]

        # Normalize the image data to 0~1
        image = (image - np.min(image)) / (np.max(image) - np.min(image))

        # Convert image data to 8-bit integers
        image = np.uint8(image*255)

        # CLAHE 적용
        clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
        clahe_img = clahe.apply(image)

        # 이미지를 PNG 파일로 저장합니다. 여기서 'i+1'은 원래 슬라이스 번호를 나타냅니다.
        plt.imsave(os.path.join(image_dir, f'OAS1_{patient_id}_{file_type}_{i+1:03}.png'), clahe_img, cmap='gray')
        
```

```python
#Methrod2
import numpy as np
import matplotlib.pyplot as plt
import glob
import os
import cv2
import math

# 함수 정의
def remove_extreme_percentiles(data, lower=10, upper=90):
    # 각 행별로 하위 및 상위 백분위수를 계산합니다.
    lower_percentiles = np.percentile(data, lower, axis=1)
    upper_percentiles = np.percentile(data, upper, axis=1)
    # 각 행별로 하위 및 상위 백분위수보다 작거나 큰 값을 찾습니다.
    mask = [(data[i] < lower_percentiles[i]) | (data[i] > upper_percentiles[i]) for i in range(data.shape[0])]
    # dtype을 명시적으로 object로 지정하여 서로 다른 길이의 배열을 저장하게 합니다.
    return np.array([row[~mask_row] for row, mask_row in zip(data, mask)], dtype=object)

def nearest_square(n):
    return math.ceil(n**0.5)**2

# 대상 디렉토리 설정
base_dir = r'C:\Users\kwoos\Desktop\cording\good\ADND\sagittal'
output_path = r'C:\Users\kwoos\Desktop\cording\good\Method2\sagittal'

# 파일 경로 내의 모든 .npy 파일을 찾습니다.
files = glob.glob(os.path.join(base_dir, '*.npy'))

# 각 .npy 파일에 대해 처리 작업을 수행합니다.
for file_idx, file in enumerate(files):
    print(f"Processing file: {file}")  # 현재 처리 중인 파일 이름 출력

    # 파일명에서 정보를 가져오기
    file_name = os.path.basename(file)  # 파일 이름과 확장자를 가져옵니다
    file_base_name = os.path.splitext(file_name)[0]  # 파일 확장자를 제거
    file_info = file_base_name.split('_')  # "_"로 나누어 정보 부분 가져오기
    patient_id = file_info[1]  # 번호 부분
    file_type = file_info[-2]  # 종류 부분(sagittal, coronal, axial)

    # 'OAS1_{번호}_{종류}' 폴더를 생성
    output_dir = os.path.join(output_path, f'OAS1_{patient_id}_{file_type}')
    os.makedirs(output_dir, exist_ok=True)

    # .npy 파일 불러오기
    data = np.load(file)

    # 결과를 저장할 리스트 초기화
    result = []

    # 반복 범위를 데이터의 인덱스 범위 내로 설정
    start = max(0, data.shape[0] // 2 - 80)
    end = min(data.shape[0], data.shape[0] // 2 + 80)

    # i 값을 start부터 end까지 바꿔가며 실행
    for i in range(start, end+1):
        # 'i'가 가운데 인덱스인 경우를 처리
        if i == data.shape[0] // 2:
            continue  # i가 가운데 인덱스일 때는 건너뛰기

        # 데이터가 4D 라면, 첫 번째 축을 따라 i번째 slice를 선택합니다.
        if len(data.shape) == 4:
            data_slice = data[i, :, :, 0]  # 4차원을 2차원으로 줄입니다.

            # 결과 리스트에 추가
            result.append(data_slice)

    # 결과를 numpy array로 변환
    result = np.array(result)

    reshaped_data = result.reshape(160, -1)
    filtered_array = remove_extreme_percentiles(reshaped_data, lower=10, upper=90)

    entropy_elements = {}
    for idx, row in enumerate(filtered_array):
        if row.size != 0:
            image = row # 현재 이미지를 선택합니다.
            side_length = int(math.sqrt(nearest_square(row.size))) #가장 가까운 제곱수를 찾습니다.
            padding = side_length**2 - row.size # 필요한 패딩의 양을 계산합니다.
            padded_row = np.pad(row, (0, padding), mode='constant') # 패딩을 추가합니다.
            image = padded_row.reshape((side_length, side_length)) # 이 행이 재구성 가능해진다.
            image = np.uint8(image) # 8-bit 정수로 변환한다.
            equalized_image = cv2.equalizeHist(image)   # OpenCV를 사용하여 히스토그램 평활화 적용
            hist = cv2.calcHist([equalized_image], [0], None, [256], [0,256]) # 히스토그램 계산
            hist /= hist.sum()  # normalize
            entropy = -np.sum(hist*np.log2(hist + np.finfo(float).eps)) # 엔트로피 계산: -sum(p(x)log(p(x)))
            entropy_elements[idx] = entropy # 결과를 딕셔너리에 저장

    # max_entropy_int = int(max(entropy_elements.values()))
    filtered_entropy_elements = {k: v for k, v in entropy_elements.items() if int(v) < 50}
    # 이제 엔트로피가 가장 높은 20개의 행의 인덱스를 추출합니다.
    # 딕셔너리를 값에 따라 내림차순으로 정렬하고, 키(인덱스)만 추출합니다.
    sorted_indices = [key for key, value in sorted(filtered_entropy_elements.items(), key=lambda item: item[1], reverse=True)]
    # 상위 20개의 행의 인덱스만 선택합니다.
    top_20_indices = sorted_indices[:20]

    # 배열에 저장된 인덱스에 해당하는 이미지를 출력하고 저장합니다.
    for idx, i in enumerate(top_20_indices):
        # 현재 이미지를 선택합니다.
        image = result[i]

        # Normalize the image data to 0~1
        image = (image - np.min(image)) / (np.max(image) - np.min(image))

        # Convert image data to 8-bit integers
        image = np.uint8(image*255)

        # CLAHE 적용
        clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
        clahe_img = clahe.apply(image)

        # 이미지를 PNG 파일로 저장합니다. 여기서 'i+1'은 원래 슬라이스 번호를 나타냅니다.
        plt.imsave(os.path.join(output_dir, f'OAS1_{patient_id}_{file_type}_{i+1:03}.png'), clahe_img, cmap='gray')
```
